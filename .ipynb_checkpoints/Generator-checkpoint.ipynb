```python
import pylab
import scipy.signal as signal
import numpy
```

```python
pqrst = signal.wavelets.daub(20)
```

```python
pqrst
```

```python
len(pqrst)
```

```python
samples_rest = 10
zero_array = numpy.zeros(samples_rest, dtype=float)
```

```python
zero_array
```

```python
pqrst_full = numpy.concatenate([pqrst,zero_array])
```

```python
pqrst_full
```

```python
pylab.plot(pqrst_full)
```
